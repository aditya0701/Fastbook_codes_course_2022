```python
from fastai.vision.all import *
import fastai
import fastcore
```

```python
path = untar_data(URLs.IMAGENETTE, archive='D:\FastAi_Data')
```

```python
dblock = DataBlock(
    blocks = (ImageBlock,CategoryBlock),
    get_items= get_image_files,
    get_y = parent_label,
    item_tfms= Resize(460),
    batch_tfms= aug_transforms(size=224, min_scale=0.75))
dls = dblock.dataloaders(path,bs=64)
```

```python
ds  = dblock.datasets(path)
```

```python
len(ds.train),len(ds.valid)
```

```python
from PIL import Image
```

```python
def show_custom_image_mit_target(idx=7):
    img = ds.train[idx]
    im_target = ds.train.vocab[idx]
    return img[0], im_target
```

```python
im,tar = show_custom_image_mit_target()
```
